```python
from tkinter import *
from tkinter import messagebox
import time
import random
```

```python
class Merge_Sort():
    def __init__(self):
        self.newWin=Tk()
        self.newWin.maxsize(900,600)
        self.newWin.configure(bg="black")
        self.newWin.title("Merge Sort")
        self.frame = Frame(self.newWin, width= 600, height=200, bg='grey')
        self.frame.grid(row=0, column=0, padx=10, pady=5)
        self.canvas = Canvas(self.newWin, width=600, height=380, bg='white')
        self.canvas.grid(row=1, column=0, padx=10, pady=5)
        Label(self.frame, text="Merge Sort : ", bg='grey').grid(row=0, column=0, padx=5, pady=5, sticky=W)
        self.speed = Scale(self.frame, from_=0.1, to=2.0, length=200, digits=2, resolution=0.2, orient=HORIZONTAL, label="Speed")
        self.speed.grid(row=0, column=2, padx=5, pady=5)
        self.array_size = Scale(self.frame, from_=3, to=25, resolution=1, orient=HORIZONTAL, label="Data Size")
        self.array_size.grid(row=1, column=0, padx=5, pady=5)
        self.min_value = Scale(self.frame, from_=0, to=10, resolution=1, orient=HORIZONTAL, label="Min Value")
        self.min_value.grid(row=1, column=1, padx=5, pady=5)
        self.max_value= Scale(self.frame, from_=10, to=100, resolution=1, orient=HORIZONTAL, label="Max Value")
        self.max_value.grid(row=1, column=2, padx=5, pady=5)
        self.data=[]
        Button(self.frame, text="Start", command=self.merge_sort, bg='#000336',fg='white').grid(row=0, column=3, padx=5, pady=5)
        Button(self.frame, text="Generate", command=self.Generate, bg='white').grid(row=1, column=3, padx=5, pady=5)
    def Generate(self):
        min_val=int(self.min_value.get())
        max_val=int(self.max_value.get())
        n=int(self.array_size.get())
        self.data=[]
        for i in range(n):
            self.data.append(random.randrange(min_val, max_val+1))
            i=i+1
        self.drawData(['red' for x in range(len(self.data))])
    def merge_sort(self):
        self.merge_sort_alg(0, len(self.data)-1)


    def merge_sort_alg(self, left, right):
        if left < right:
            middle = (left + right) // 2
            self.merge_sort_alg(left, middle)
            self.merge_sort_alg(middle+1, right)
            self.merge(left, middle, right)

    def merge(self, left, middle, right):
        self.drawData(self.getColorArrayb(len(self.data), left, middle, right))
        time.sleep(self.speed.get())
        leftPart = self.data[left:middle+1]
        rightPart = self.data[middle+1: right+1]
        leftIdx = rightIdx = 0
        for dataIdx in range(left, right+1):
            if leftIdx < len(leftPart) and rightIdx < len(rightPart):
                if leftPart[leftIdx] <= rightPart[rightIdx]:
                    self.data[dataIdx] = leftPart[leftIdx]
                    leftIdx += 1
                else:
                    self.data[dataIdx] = rightPart[rightIdx]
                    rightIdx += 1
        
            elif leftIdx < len(leftPart):
                self.data[dataIdx] = leftPart[leftIdx]
                leftIdx += 1
            else:
                self.data[dataIdx] = rightPart[rightIdx]
                rightIdx += 1
    
        self.drawData(["green" if x >= left and x <= right else "white" for x in range(len(self.data))])
        time.sleep(self.speed.get())

    def getColorArrayb(self,leght, left, middle, right):
        colorArray = []

        for i in range(leght):
            if i >= left and i <= right:
                if i >= left and i <= middle:
                    colorArray.append("yellow")
                else:
                    colorArray.append("pink")
            else:
                colorArray.append("white")

        return colorArray
    
    def drawData(self,colorArray):
        self.canvas.delete("all")
        c_height = 380
        c_width = 600
        x_width = c_width / (len(self.data) + 1)
        offset = 30
        spacing = 10
        normalizedData = [ i / max(self.data) for i in self.data]
        for i, height in enumerate(normalizedData):
            #top left
            x0 = i * x_width + offset + spacing
            y0 = c_height - height * 340
            #bottom right
            x1 = (i + 1) * x_width + offset
            y1 = c_height

            self.canvas.create_rectangle(x0, y0, x1, y1, fill=colorArray[i])
            self.canvas.create_text(x0+2, y0, anchor=SW, text=str(self.data[i]))
    
        self.newWin.update_idletasks()
    def run(self):
        self.newWin.mainloop()
```

```python
def open_mergesort():
    m=Merge_Sort()
    m.run()
```

```python
class Bubble_Sort():
    def __init__(self):
        self.newWin=Tk()
        self.newWin.maxsize(900,600)
        self.newWin.configure(bg="black")
        self.newWin.title("Bubble Sort")
        self.frame = Frame(self.newWin, width= 600, height=200, bg='grey')
        self.frame.grid(row=0, column=0, padx=10, pady=5)
        self.canvas = Canvas(self.newWin, width=600, height=380, bg='white')
        self.canvas.grid(row=1, column=0, padx=10, pady=5)
        Label(self.frame, text="Bubble Sort : ", bg='grey').grid(row=0, column=0, padx=5, pady=5, sticky=W)
        self.speed = Scale(self.frame, from_=0.1, to=2.0, length=200, digits=2, resolution=0.2, orient=HORIZONTAL, label="Speed")
        self.speed.grid(row=0, column=2, padx=5, pady=5)
        self.array_size = Scale(self.frame, from_=3, to=25, resolution=1, orient=HORIZONTAL, label="Data Size")
        self.array_size.grid(row=1, column=0, padx=5, pady=5)
        self.min_value = Scale(self.frame, from_=0, to=10, resolution=1, orient=HORIZONTAL, label="Min Value")
        self.min_value.grid(row=1, column=1, padx=5, pady=5)
        self.max_value= Scale(self.frame, from_=10, to=100, resolution=1, orient=HORIZONTAL, label="Max Value")
        self.max_value.grid(row=1, column=2, padx=5, pady=5)
        self.data=[]
        Button(self.frame, text="Start", command=self.bubble, bg='#000336',fg='white').grid(row=0, column=3, padx=5, pady=5)
        Button(self.frame, text="Generate", command=self.Generate, bg='white').grid(row=1, column=3, padx=5, pady=5)
    def Generate(self):
        min_val=int(self.min_value.get())
        max_val=int(self.max_value.get())
        n=int(self.array_size.get())
        self.data=[]
        for i in range(n):
            self.data.append(random.randrange(min_val, max_val+1))
            i=i+1
        self.drawData(['red' for x in range(len(self.data))])
    def bubble_sort(self):
        for _ in range(len(self.data)-1):
            for j in range(len(self.data)-1):
                if self.data[j] > self.data[j+1]:
                    self.data[j], self.data[j+1] = self.data[j+1], self.data[j]
                    self.drawData(['yellow' if x == j or x == j+1 else 'red' for x in range(len(self.data))] )
                    time.sleep(self.speed.get())
    def bubble(self):
        self.bubble_sort()
        self.drawData(['green' for x in range(len(self.data))])
    
    def drawData(self,colorArray):
        self.canvas.delete("all")
        c_height = 380
        c_width = 600
        x_width = c_width / (len(self.data) + 1)
        offset = 30
        spacing = 10
        normalizedData = [ i / max(self.data) for i in self.data]
        for i, height in enumerate(normalizedData):
            #top left
            x0 = i * x_width + offset + spacing
            y0 = c_height - height * 340
            #bottom right
            x1 = (i + 1) * x_width + offset
            y1 = c_height

            self.canvas.create_rectangle(x0, y0, x1, y1, fill=colorArray[i])
            self.canvas.create_text(x0+2, y0, anchor=SW, text=str(self.data[i]))
    
        self.newWin.update_idletasks()
    def run(self):
        self.newWin.mainloop()
```

```python
def open_bubblesort():
    print("bubble sort")
    b=Bubble_Sort()
    b.run()
```

```python
class Insertion_Sort():
    def __init__(self):
        self.newWin=Tk()
        self.newWin.maxsize(900,600)
        self.newWin.configure(bg="black")
        self.newWin.title("insertion_sort")
        self.frame = Frame(self.newWin, width= 600, height=200, bg='grey')
        self.frame.grid(row=0, column=0, padx=10, pady=5)
        self.canvas = Canvas(self.newWin, width=600, height=380, bg='white')
        self.canvas.grid(row=1, column=0, padx=10, pady=5)
        Label(self.frame, text="Bubble Sort : ", bg='grey').grid(row=0, column=0, padx=5, pady=5, sticky=W)
        self.speed = Scale(self.frame, from_=0.1, to=2.0, length=200, digits=2, resolution=0.2, orient=HORIZONTAL, label="Speed")
        self.speed.grid(row=0, column=2, padx=5, pady=5)
        self.array_size = Scale(self.frame, from_=3, to=25, resolution=1, orient=HORIZONTAL, label="Data Size")
        self.array_size.grid(row=1, column=0, padx=5, pady=5)
        self.min_value = Scale(self.frame, from_=0, to=10, resolution=1, orient=HORIZONTAL, label="Min Value")
        self.min_value.grid(row=1, column=1, padx=5, pady=5)
        self.max_value= Scale(self.frame, from_=10, to=100, resolution=1, orient=HORIZONTAL, label="Max Value")
        self.max_value.grid(row=1, column=2, padx=5, pady=5)
        self.data=[]
        Button(self.frame, text="Start", command=self.insertion_sort, bg='#000336',fg='white').grid(row=0, column=3, padx=5, pady=5)
        Button(self.frame, text="Generate", command=self.generate, bg='white').grid(row=1, column=3, padx=5, pady=5)
    def generate(self):
        min_val=int(self.min_value.get())
        max_val=int(self.max_value.get())
        n=int(self.array_size.get())
        self.data=[]
        for i in range(n):
            self.data.append(random.randrange(min_val, max_val+1))
            i=i+1
        self.drawData(['red' for x in range(len(self.data))])
    def insertion_sort(self):
        colorData=['red' for x in range(len(self.data))]
        for i in range(1,len(self.data)):
            key=self.data[i]
            j=i-1
            colorData[i]='blue'
            self.drawData(colorData)
            time.sleep(self.speed.get())
            while j>=0 and self.data[j]>key:
                colorData[j]='white'
                self.drawData(colorData)
                time.sleep(self.speed.get())
                colorData[j]='red'
                self.data[j+1]=self.data[j]
                j-=1
            self.data[j+1]=key

            colorData[j+1]='grey'
            self.drawData(colorData)
            time.sleep(self.speed.get())
            colorData[j+1]='red'
            colorData[i]='red'
        colorData=['green' for x in range(len(self.data))]
        self.drawData(colorData)
    def drawData(self,colorArray):
        self.canvas.delete("all")
        c_height = 380
        c_width = 600
        x_width = c_width / (len(self.data) + 1)
        offset = 30
        spacing = 10
        normalizedData = [ i / max(self.data) for i in self.data]
        for i, height in enumerate(normalizedData):
            #top left
            x0 = i * x_width + offset + spacing
            y0 = c_height - height * 340
            #bottom right
            x1 = (i + 1) * x_width + offset
            y1 = c_height

            self.canvas.create_rectangle(x0, y0, x1, y1, fill=colorArray[i])
            self.canvas.create_text(x0+2, y0, anchor=SW, text=str(self.data[i]))
    
        self.newWin.update_idletasks()        
    def run(self):
        self.newWin.mainloop()    
```

```python
class Quick_Sort():
    def __init__(self):
        self.newWin=Tk()
        self.newWin.maxsize(900,600)
        self.newWin.configure(bg="black")
        self.newWin.title("Quick Sort")
        self.frame = Frame(self.newWin, width= 600, height=200, bg='grey')
        self.frame.grid(row=0, column=0, padx=10, pady=5)
        self.canvas = Canvas(self.newWin, width=600, height=380, bg='white')
        self.canvas.grid(row=1, column=0, padx=10, pady=5)
        Label(self.frame, text="Quick Sort : ", bg='grey').grid(row=0, column=0, padx=5, pady=5, sticky=W)
        self.speed = Scale(self.frame, from_=0.1, to=2.0, length=200, digits=2, resolution=0.2, orient=HORIZONTAL, label="Speed")
        self.speed.grid(row=0, column=2, padx=5, pady=5)
        self.array_size = Scale(self.frame, from_=3, to=25, resolution=1, orient=HORIZONTAL, label="Data Size")
        self.array_size.grid(row=1, column=0, padx=5, pady=5)
        self.min_value = Scale(self.frame, from_=0, to=10, resolution=1, orient=HORIZONTAL, label="Min Value")
        self.min_value.grid(row=1, column=1, padx=5, pady=5)
        self.max_value= Scale(self.frame, from_=10, to=100, resolution=1, orient=HORIZONTAL, label="Max Value")
        self.max_value.grid(row=1, column=2, padx=5, pady=5)
        self.data=[]
        Button(self.frame, text="Start", command=self.quick, bg='#000336',fg='white').grid(row=0, column=3, padx=5, pady=5)
        Button(self.frame, text="Generate", command=self.Generate, bg='white').grid(row=1, column=3, padx=5, pady=5)
    def Generate(self):
        min_val=int(self.min_value.get())
        max_val=int(self.max_value.get())
        n=int(self.array_size.get())
        self.data=[]
        for i in range(n):
            self.data.append(random.randrange(min_val, max_val+1))
            i=i+1
        self.drawData(['red' for x in range(len(self.data))])
    def partition(self, head, tail):
        border = head
        pivot = self.data[tail]
        self.drawData(self.getColorArray(len(self.data), head, tail, border, border))
        time.sleep(self.speed.get())
        for j in range(head, tail):
            if self.data[j] < pivot:
                self.drawData(self.getColorArray(len(self.data), head, tail, border, j, True))
                time.sleep(self.speed.get())
                self.data[border], self.data[j] = self.data[j], self.data[border]
                border += 1
            self.drawData(self.getColorArray(len(self.data), head, tail, border, j))
            time.sleep(self.speed.get())


    #swap pivot with border value
        self.drawData(self.getColorArray(len(self.data), head, tail, border, tail, True))
        time.sleep(self.speed.get())

        self.data[border], self.data[tail] = self.data[tail], self.data[border]
    
        return border

    def quick_sort(self, head, tail):
        if head < tail:
            partitionIdx = self.partition(head, tail)

            #LEFT PARTITION
            self.quick_sort(head, partitionIdx-1)

            #RIGHT PARTITION
            self.quick_sort(partitionIdx+1, tail)

    def quick(self):
        self.quick_sort(0,len(self.data)-1)
        self.drawData( ['green' for x in range(len(self.data))])
    def getColorArray(self,dataLen, head, tail, border, currIdx, isSwaping = False):
        colorArray = []
        for i in range(dataLen):
            #base coloring
            if i >= head and i <= tail:
                colorArray.append('gray')
            else:
                colorArray.append('white')
            if i == tail:
                colorArray[i] = 'blue'
            elif i == border:
                colorArray[i] = 'red'
            elif i == currIdx:
                colorArray[i] = 'yellow'
            if isSwaping:
                if i == border or i == currIdx:
                    colorArray[i] = 'green'

        return colorArray
    
    def drawData(self,colorArray):
        self.canvas.delete("all")
        c_height = 380
        c_width = 600
        x_width = c_width / (len(self.data) + 1)
        offset = 30
        spacing = 10
        normalizedData = [ i / max(self.data) for i in self.data]
        for i, height in enumerate(normalizedData):
            #top left
            x0 = i * x_width + offset + spacing
            y0 = c_height - height * 340
            #bottom right
            x1 = (i + 1) * x_width + offset
            y1 = c_height

            self.canvas.create_rectangle(x0, y0, x1, y1, fill=colorArray[i])
            self.canvas.create_text(x0+2, y0, anchor=SW, text=str(self.data[i]))
    
        self.newWin.update_idletasks()
    def run(self):
        self.newWin.mainloop()
```

```python
class Selection_Sort():
    def __init__(self):
        self.newWin=Tk()
        self.newWin.maxsize(900,600)
        self.newWin.configure(bg="black")
        self.newWin.title("selection_sort")
        self.frame = Frame(self.newWin, width= 600, height=200, bg='grey')
        self.frame.grid(row=0, column=0, padx=10, pady=5)
        self.canvas = Canvas(self.newWin, width=600, height=380, bg='white')
        self.canvas.grid(row=1, column=0, padx=10, pady=5)
        Label(self.frame, text="Bubble Sort : ", bg='grey').grid(row=0, column=0, padx=5, pady=5, sticky=W)
        self.speed = Scale(self.frame, from_=0.1, to=2.0, length=200, digits=2, resolution=0.2, orient=HORIZONTAL, label="Speed")
        self.speed.grid(row=0, column=2, padx=5, pady=5)
        self.array_size = Scale(self.frame, from_=3, to=25, resolution=1, orient=HORIZONTAL, label="Data Size")
        self.array_size.grid(row=1, column=0, padx=5, pady=5)
        self.min_value = Scale(self.frame, from_=0, to=10, resolution=1, orient=HORIZONTAL, label="Min Value")
        self.min_value.grid(row=1, column=1, padx=5, pady=5)
        self.max_value= Scale(self.frame, from_=10, to=100, resolution=1, orient=HORIZONTAL, label="Max Value")
        self.max_value.grid(row=1, column=2, padx=5, pady=5)
        self.data=[]
        Button(self.frame, text="Start", command=self.selection_sort, bg='#000336',fg='white').grid(row=0, column=3, padx=5, pady=5)
        Button(self.frame, text="Generate", command=self.generate, bg='white').grid(row=1, column=3, padx=5, pady=5)
    def generate(self):
        min_val=int(self.min_value.get())
        max_val=int(self.max_value.get())
        n=int(self.array_size.get())
        self.data=[]
        for i in range(n):
            self.data.append(random.randrange(min_val, max_val+1))
            i=i+1
        self.drawData(['red' for x in range(len(self.data))])
    def selection_sort(self):
        colorData=['red' for x in range(len(self.data))]
        for i in range(len(self.data)):
            m=self.data[i]
            min_index=i
            colorData[i]='blue'
            for j in range(i+1,len(self.data)):
                colorData[j]='white'
                self.drawData(colorData)
                colorData[j]='red'
                time.sleep(self.speed.get())
                if self.data[j]<m:
                    colorData[min_index]='red'
                    m=self.data[j]
                    min_index=j
                    colorData[j]='blue'
                    if j!=len(self.data)-1:
                        colorData[j+1]='white'
                    self.drawData(colorData)
                    # colorData[j]='blue'
                    time.sleep(self.speed.get())
            self.data[min_index],self.data[i]=self.data[i],self.data[min_index]
            colorData[i+1:]=['red' for x in range(i+1,len(self.data))]
            colorData[i]='green'
            self.drawData(colorData)
        time.sleep(self.speed.get())
    def drawData(self,colorArray):
        self.canvas.delete("all")
        c_height = 380
        c_width = 600
        x_width = c_width / (len(self.data) + 1)
        offset = 30
        spacing = 10
        normalizedData = [ i / max(self.data) for i in self.data]
        for i, height in enumerate(normalizedData):
            #top left
            x0 = i * x_width + offset + spacing
            y0 = c_height - height * 340
            #bottom right
            x1 = (i + 1) * x_width + offset
            y1 = c_height

            self.canvas.create_rectangle(x0, y0, x1, y1, fill=colorArray[i])
            self.canvas.create_text(x0+2, y0, anchor=SW, text=str(self.data[i]))
    
        self.newWin.update_idletasks()        
    def run(self):
        self.newWin.mainloop()        
```
